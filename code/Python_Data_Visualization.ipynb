```python
# scatter plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)

#  Matplotlib 
plt.scatter(x, y, marker="x")
plt.show()

#  seaborn 
df = pd.DataFrame({'x':x, 'y':y})
sns.jointplot(x="x", y="y", data=df, kind="scatter")
plt.show()
```

```python
# line plot,
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]

#  Matplotlib 
plt.plot(x, y)
plt.show()

#  Seaborn 
df = pd.DataFrame({'year':x, 'num':y})
sns.lineplot(x="year", y="num", data=df)
plt.show()
```

```python
# dist plot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

#  0 ~ 100  100 
s = np.random.randint(0, 100, 100)
bins = np.arange(0, 101, 10)

print(s)

#  Matplotlib 
plt.hist(s, bins)
plt.show()

#  seaborn 
sns.distplot(s, bins=10, kde=False)
plt.show()
sns.distplot(s, bins=10, kde=True)
plt.show()
```

```python
# 
# 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

#  0 - 1  10*4 
data = np.random.normal(size=(10, 5))
labels = ["A", "B", "C", "D", "E"]

print(data)

#  Matplotlib 
plt.boxplot(data, labels=labels)
plt.show()

#  Seaborn 
df = pd.DataFrame(data, columns=labels)
sns.boxplot(data=df)
plt.show()
```

```python
# heat map
# Seaborn  1949  1960 
import seaborn as sns
import matplotlib.pyplot as plt

# 
flights = sns.load_dataset("flights")
data = flights.pivot('year', 'month', 'passengers')

print(flights)
print("--------------------------------------------------------------------")
print(data)

#  Seaborn 
sns.heatmap(data)
plt.show()
```

```python
# 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties

# 
labels = np.array([u"  ", " KDA ", u"  ", u"  ", u"  ", u"  "])
stats = [83, 61, 95, 67, 76, 88]

# 

# 
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

# 
stats = np.concatenate((stats, [stats[0]]))
angles = np.concatenate((angles, [angles[0]]))

#  Matplotlib 
fig = plt.figure()

# 1  1 1 
ax = fig.add_subplot(111, polar=True)

# 
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)

# 
font = FontProperties(fname=r"C:/Windows/Fonts/simhei.ttf", size=14)
ax.set_thetagrids(angles * 180 / np.pi, labels, FontProperties=font)

plt.show()
```

```python
# 
import matplotlib.pyplot as plt
import seaborn as sns


# total_bill tip 
tips = sns.load_dataset("tips")
print(tips.head(10))

#  Seaborn Hexbin 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```
