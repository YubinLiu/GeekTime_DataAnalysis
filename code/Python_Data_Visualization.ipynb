```python
# 散点图（scatter plot），展示两个变量之间的关系
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)

# 用 Matplotlib 画散点图
plt.scatter(x, y, marker="x")
plt.show()

# 用 seaborn 画散点图
df = pd.DataFrame({'x':x, 'y':y})
sns.jointplot(x="x", y="y", data=df, kind="scatter")
plt.show()
```

```python
# 折线图（line plot）,数据变化趋势
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 数据准备
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]

# 使用 Matplotlib 画折线图
plt.plot(x, y)
plt.show()

# 使用 Seaborn 画折线图
df = pd.DataFrame({'year':x, 'num':y})
sns.lineplot(x="year", y="num", data=df)
plt.show()
```

```python
# 直方图（dist plot），把横坐标分成一定数量的小区间（“箱子”）
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 数据准备，生成 0 ~ 100 之间的 100 个数据
s = np.random.randint(0, 100, 100)
bins = np.arange(0, 101, 10)

print(s)

# 用 Matplotlib 画直方图
plt.hist(s, bins)
plt.show()

# 用 seaborn 画直方图
sns.distplot(s, bins=10, kde=False)
plt.show()
sns.distplot(s, bins=10, kde=True)
plt.show()
```

```python
# 条形图（bar plot），查看类别的频数
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
x = ["Cat1", "Cat2", "Cat3", "Cat4", "Cat5"]
y = [5, 4, 8, 12, 7]

# 用 Matplotlib 画条形图
plt.bar(x, y)
plt.show()

# 用 Seaborn 画条形图
sns.barplot(x, y)
plt.show()
```

```python
# 箱线图（盒式图），最大值、最小值、中位数、上下四分位数。
# 分析数据差异性、离散性和异常值。
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 数据准备，生成 0 - 1 之间的 10*4 维度数据
data = np.random.normal(size=(10, 5))
labels = ["A", "B", "C", "D", "E"]

print(data)

# 使用 Matplotlib 画箱线图
plt.boxplot(data, labels=labels)
plt.show()

# 使用 Seaborn 画箱线图
df = pd.DataFrame(data, columns=labels)
sns.boxplot(data=df)
plt.show()
```

```python
# 热力图（heat map），矩阵表示方法，矩阵的元素值用颜色来代表
# Seaborn 自带数据集，记录了 1949 年到 1960 年之间，每个月的航班乘客数量。
import seaborn as sns
import matplotlib.pyplot as plt

# 数据准备
flights = sns.load_dataset("flights")
data = flights.pivot('year', 'month', 'passengers')

print(flights)
print("--------------------------------------------------------------------")
print(data)

# 用 Seaborn 画热力图
sns.heatmap(data)
plt.show()
```

```python
# 蜘蛛图，显示一对多的关系
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties

# 数据准备
labels = np.array([u" 推进 ", " KDA ", u" 生存 ", u" 团战 ", u" 发育 ", u" 输出 "])
stats = [83, 61, 95, 67, 76, 88]

# 画图数据准备，角度、状态值

# 设置雷达图的角度，用于平分切开一个圆面
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

# 为了使雷达图封闭，数组拼接
stats = np.concatenate((stats, [stats[0]]))
angles = np.concatenate((angles, [angles[0]]))

# 用 Matplotlib 画蜘蛛图
fig = plt.figure()

# 1 行 1列 1 个子图，极坐标格式
ax = fig.add_subplot(111, polar=True)

# 画折线图
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)

# 设置中文字体
font = FontProperties(fname=r"C:/Windows/Fonts/simhei.ttf", size=14)
ax.set_thetagrids(angles * 180 / np.pi, labels, FontProperties=font)

plt.show()
```

```python
# 二元变量分布，探究两个变量之间的关系
import matplotlib.pyplot as plt
import seaborn as sns


# 数据准备，total_bill 客户账单金额，tip 小费金额
tips = sns.load_dataset("tips")
print(tips.head(10))

# 用 Seaborn 画二元变量分布图（散点图、核密度图、Hexbin 图）
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```

```python
# 成对关系，探索数据集中多个成对变量的分布
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
iris = sns.load_dataset("iris")

# 用 Seaborn 画成对关系
sns.pairplot(iris)
plt.show()
```
