```python
import pandas as pd

df = pd.read_excel("./data/服装店数据.xlsx")

df.dropna(how="all", inplace=True)

# Weight 列数据单位不统一，统一转换为 kgs
# 2.21 lbs = 1 kgs
rows_with_lbs = df["Weight"].str.contains("lbs").fillna(False)
df[rows_with_lbs]

for i, lbs_row in df[rows_with_lbs].iterrows():
    weight = int(float(lbs_row["Weight"][:-3])/2.2)
    df.at[i, "Weight"] = '{}kgs'.format(weight)

df
```

```python
# Name 拆分成 FirstName 和 LastName
df[["FirstName", "LastName"]] = df["Name"].str.split(expand=True)
df.drop("Name", axis=1, inplace=True)
```

```python
first_name = df["FirstName"]
last_name = df["LastName"]
df.drop(labels=["FirstName", "LastName"], axis=1, inplace=True)
```

```python
df.insert(0, "FirstName", first_name)
df.insert(1, "LastName", last_name)
```

```python
df
```

```python
# 删除重复行
df.drop_duplicates(["FirstName", "LastName"], inplace=True)
```

```python
df
```

```python
# 平均值填充 Age
df["Age"].fillna(df['Age'].mean(), inplace=True)
df.round(1)
```
